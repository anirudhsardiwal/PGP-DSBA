```python
df['Product'].value_counts().head(10).plot(kind='bar');
plt.xlabel('Product', fontsize = 12);
plt.ylabel('Count', fontsize=12);
```

```python
df.resample('W', on="Date").nunique()['Order_id'].plot(figsize=(18,3));
plt.xlabel('Order Date (Weekly)', fontsize = 14);
plt.ylabel('Count of Orders', fontsize=14);
plt.xticks(rotation=0, fontsize=12);
plt.yticks(fontsize=12);
```

```python
df.resample('ME', on="Date").nunique()['Order_id'].plot(figsize=(18,3), marker='o');
plt.xlabel('Order Date (Monthly)', fontsize = 14);
plt.ylabel('Count of Orders', fontsize=14);
plt.xticks(rotation=0, fontsize=12);
plt.yticks(fontsize=12);
```

```python
df.resample('QE', on="Date").nunique()['Order_id'].plot(figsize=(18,3), marker='o');
plt.xlabel('Order Date (Quarterly)', fontsize = 14);
plt.ylabel('Count of Orders', fontsize=14);
plt.xticks(rotation=0, fontsize=12);
plt.yticks(fontsize=12);
```
